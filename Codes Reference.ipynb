# Codes Reference
This document provides reference material about the codes included in our dataset, namely listing the codes and their associated descriptions. We proceed for shortest number of codes to largest.

```python
# imports and data loading
import pandas as pd
pd.set_option("display.max_rows", 500)
```

```python
full_dataset = pd.read_csv("data/Hospital_Inpatient_Discharges__SPARCS_De-Identified___2015.csv")
```

```python
codes = [
    ("CCS Diagnosis Code", "CCS Diagnosis Description"),
    ("CCS Procedure Code", "CCS Procedure Description"),
    ("APR DRG Code", "APR DRG Description"),
    ("APR MDC Code", "APR MDC Description"),
    ("APR Severity of Illness Code", "APR Severity of Illness Description"),
]

for code in codes:
    print(code)
    print(full_dataset[list(code)].nunique())
```

```python
index = -1
print("Severity of Illness")
full_dataset[list(codes[index])].drop_duplicates().set_index(codes[index][0]).sort_index()
```

```python
index = -2
print("Major Diagnostic Category")
full_dataset[list(codes[index])].drop_duplicates().set_index(codes[index][0]).sort_index()
```
